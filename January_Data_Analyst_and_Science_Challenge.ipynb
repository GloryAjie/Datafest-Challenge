#Now to solve the problem we need to calculate the AOV for distinct sneakers using shopId since each shop has a unique sneaker


```python
grouper = df.groupby(["shop_id"])
(
    (grouper['total_items'].sum() / grouper['order_amount'].nunique())
    .reset_index(name = 'distinctAverage')
)

```
